```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme()
```

```python
df_qiskit_qikist_ye_ye = pd.read_csv("../../qlbm_output/qiskit_qiskit/data_simulation.csv")
df_qiskit_qikist_no_ye = pd.read_csv("../../qlbm_output/qiskit_qiskit_no_ye/data_simulation.csv")
df_tket_qikist_ye_ye = pd.read_csv("../../qlbm_output/tket_qiskit/data_simulation.csv")

df_qiskit_qikist_ye_ye
```

```python
plt.clf()

sns.lineplot(data=df_qiskit_qikist_ye_ye, x="Step", y="Simulation Duration", label = "Qiskit Compiler, Snapshots")
sns.lineplot(data=df_qiskit_qikist_no_ye, x="Step", y="Simulation Duration", label = "Qiskit Compiler, No Snapshots")
sns.lineplot(data=df_tket_qikist_ye_ye, x="Step", y="Simulation Duration", label = "Tket Compiler, Snapshots")

plt.savefig("runtime_simulation.png", format="png")
plt.savefig("runtime_simulation.pdf", format="png")
```

```python
plt.clf()

sns.lineplot(data=df_qiskit_qikist_ye_ye, x="Step", y="Depth", label = "Qiskit Compiler, Snapshots")
sns.lineplot(data=df_qiskit_qikist_no_ye, x="Step", y="Depth", label = "Qiskit Compiler, No Snapshots")
sns.lineplot(data=df_tket_qikist_ye_ye, x="Step", y="Depth", label = "Tket Compiler, Snapshots")

plt.savefig("depth.png", format="png")
plt.savefig("depth.png", format="png")
```
